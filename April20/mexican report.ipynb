```python
import PyPDF2
import spacy
import seaborn as sns
import pandas as pd
import geopandas
import sys
import csv
from  matplotlib import pyplot as plt
import numpy as np
```

```python
reader = PyPDF2.PdfFileReader("../data/informe.pdf")
```

```python
# Data Clean
```

```python
pdf_page_number = 3
full_text=""
for i in range(14, 327):
    if pdf_page_number<=9:
        page_text = reader.getPage(i).extractText().strip()[1:]
    elif pdf_page_number>=10 and pdf_page_number<=99:
        page_text = reader.getPage(i).extractText().strip()[2:]
    else:
        page_text = reader.getPage(i).extractText().strip()[3:]
    
    full_text+=page_text.replace("\n","")
    pdf_page_number +=1
        
```

```python
CHARACTERS = {
    "ç": "Á",
    "⁄": "á",
    "…": "É",
    "”": "é",
    "ê": "Í",
    "™": 'í',
    "î": "Ó"
}
```

```python
for item, replacement in CHARACTERS.items():
    full_text= full_text.replace(item, replacement)

full_text = full_text.replace("  "," ").replace("  "," ").replace("  "," ")

with open("../data/trascript_clean.txt","w", encoding ="utf-8") as temp_file:
    temp_file.write(full_text)
```

```python
# nlp pipeline
# run python -m spacy download es_core_news_md -- spanish model
```

```python
corpus = open("../data/trascript_clean.txt","r", encoding ="utf-8").read()
nlp = spacy.load("es_core_news_md")

nlp.max_length = len(corpus)

doc = nlp(corpus)
```

```python
# text tokens
data_list = [["text","text_lower","lemma","lemma_lower","part_of_speech","is_alphabet","is_stopword"]]

for token in doc:
    data_list.append([token.text, token.lower_,token.lemma_,token.lemma_.lower(), token.pos_, token.is_alpha, token.is_stop])
    
    
csv.writer(open("./tokens.csv","w",encoding="utf-8",newline="")).writerows(data_list)
```

```python
# text entities

data_list = [["text","text_lower","label"]]

for ent in doc.ents:
    data_list.append([ent.text,ent.lower_,ent.label])
    
csv.writer(open("./entities.csv","w",encoding="utf-8",newline = "")).writerows(data_list)
```

```python
# download positive and negative lexicons from kaggle
# https://www.kaggle.com/rtatman/sentiment-lexicons-for-81-languages

with open("../data/positive_words_es.txt","r",encoding="utf-8") as temp_file:
    positive_words = temp_file.read().splitlines()
    
with open("../data/negative_words_es.txt","r", encoding="utf-8") as temp_file:
    negative_words = temp_file.read().splitlines()
    
    
data_list = [["text","score"]]

for sent in doc.sents:
    if len(sent.text)>10:
        score = 0
        
        for word in sent:
            if word.lower_ in positive_words:
                score +=1
            if word.lower_ in negative_words:
                score-=1
                
            data_list.append([sent.text,score])
            
csv.writer(open("./sentences.csv","w",encoding="utf-8",newline="")).writerows(data_list)
```

```python
df = pd.read_csv("../data/entities.csv")
mexico_df = geopandas.read_file("../data/mexstates")
```

```python
mexico_df.head()
```

```python
for state in mexico_df["ADMIN_NAME"]:
    for item, replacement in CHARACTERS.items():
        clean_name= state.replace(item, replacement)
        if clean_name == "Ciudad de Mexico":
            clean_name = "Distrito Federal"
        elif clean_name == "Estado de Mexico":
            clean_name = "Mexico"
        
        mexico_df.loc[mexico_df["ADMIN_NAME"]==clean_name,"count"]=len(df[df["text_lower"]==state.lower()])
   
```

```python
mexico_df.head()
```

```python
mexico_df.plot(column="count", cmap="plasma", legend=True)
plt.title("Mentions by state")
plt.axis("off")
plt.tight_layout()
plt.show()
```
